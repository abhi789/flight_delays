Now that we have a few insights into the data, we can proceed with cleaning the data

```python
# cleaning the data
flights_df.info()
```

```python
# lets look at the missing values
flights_df.isna().sum()
```

Flights with missing tail numbers are generally clandestine or operated by certain governments. We can freely ignore those flights as those records might not be reliable

```python
# dropping all the records with tail no missing
flights_df.dropna(subset = ["TAIL_NUMBER"], inplace = True)
```

```python
# confirming it worked
flights_df.isna().sum()
```

We can't really use the records with missing departure time either. Need to figure out a clear strategy for these records

```python
# percent of missing values
flights_df.DEPARTURE_TIME.isna().sum() / len(flights_df)
```

As the no of missing values is less than 1.5% we can freely drop all the na values without worrying about losing any potential data.

```python
flights_df.dropna(inplace = True)
```

```python
# checking that it worked
flights_df.isna().sum()
```

```python
# checking the shape
flights_df.shape
```

```python
# determing the continuos and categorical variables
cont_vars = []
cat_vars = []

for col in flights_df.columns: 
    if flights_df[col].dtype == 'O':
        cat_vars.append(col)
    else: 
        cont_vars.append(col)

print(f'There are {len(cat_vars)} categorical variables: \n', cat_vars)
print('----' * 8)
print(f'There are {len(cont_vars)} continuous variables: \n', cont_vars)
```

```python
# check the number of unique values and the percent each value represents
# for each categoritcal variable
for v in cat_vars:
    print(f'Unique values for {v}: {flights_df[v].nunique()}\n', 
          round(flights_df[v].value_counts(normalize=True)* 100, 2), '\n')
```

```python
cont_data = flights_df[cont_vars]
```

```python
cont_data.head()
```

```python
# create a function that iterates through the dataframe and plots histograms for our continuous variables 
def plot_cont(df):
    
    fig = plt.figure(figsize=(13,40))
    plt.subplots_adjust(hspace=1)

    nrows = len(df.columns) //2 +1 
    ncols = 2
    
    for i,t in enumerate(list(df.columns)):
        ax = fig.add_subplot(nrows, ncols, i+1)
        sns.distplot(a= df[t],   
                     ax=ax)
        plt.title(t + '_distribution')
```

```python
plot_cont(cont_data)
```

```python
# importing modelling libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
```

```python
flights_model = flights_df
```

```python
flights_model.info()
```

```python
flights_model = flights_model.iloc[:,:25]
```

```python
flights_model.info()
```

```python
le = LabelEncoder()
```

# importing modelling libraries
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

```python
# doing some label encoding
flights_model['AIRLINE']= le.fit_transform(flights_model['AIRLINE'])
flights_model['ORIGIN_AIRPORT'] = le.fit_transform(flights_model['ORIGIN_AIRPORT'])
flights_model['DESTINATION_AIRPORT'] = le.fit_transform(flights_model['DESTINATION_AIRPORT'])
flights_model['DAY'] = le.fit_transform(flights_model['DAY'])
```

```python
flights_model.info()
```

```python
y = flights_model["ARRIVAL_DELAY"]
```

```python
X = flights_model.drop(columns = ["ARRIVAL_DELAY", "TAIL_NUMBER"])
```

```python
X.info()
```

```python
# Splitting into train and test data set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 2)
```

```python
sc1 = StandardScaler()
X_train_sc = sc1.fit_transform(X_train)
X_test_sc = sc1.transform(X_test)
```

```python
linreg = LinearRegression()
random_forest_regressor = RandomForestRegressor()
```

```python
model_linreg = linreg.fit(X_train_sc, y_train)
model_rf = random_forest_regressor.fit(X_train_sc, y_train)
```

```python
#y_predict_linreg = model_linreg.predict(X_test_sc)
y_predict_rf = model_rf.predict(X_test_sc)
```

```python
print('Root Mean Squared Error of Linear Regression:', np.sqrt(mean_squared_error(y_test, y_predict_linreg)))
print('R2 of linear regression : ',r2_score(y_test, y_predict_linreg))
```

```python
print('Root Mean Squared Error of Random Forest Regressor:', np.sqrt(mean_squared_error(y_test, y_predict_rf)))
print('R2 of Random  Forest Regressor: ',r2_score(y_test, y_predict_rf))
```

```python
joblib.dump(model_linreg, 'linreg.pkl')
joblib.dump(model_rf, 'rf.pkl')
```
